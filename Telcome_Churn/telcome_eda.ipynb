#### Too many rows to Analyze.
##### Check Missing data and remove who have too many missing values in column

```python
missing_column = churn_data.columns[(100*churn_data.isnull().sum()/len(churn_data.index)) < 50]
churn_data=churn_data[missing_column]
churn_data.head()
```

```python
missing_column = churn_data.columns[(100*churn_data.isnull().sum()/len(churn_data.index)) > 5]
missing_column
```

#### Lots of column having more than 5% data is mssing.

```python
no_missing_churn_df = churn_data.dropna(axis=0,how='any')
no_missing_churn_df.shape
```

#### Removed rows which having any missing values

```python
no_missing_churn_df.select_dtypes(exclude=['int', 'float']).columns
```

```python
date_attributes = ['last_date_of_month_6', 'last_date_of_month_7',
       'last_date_of_month_8', 'last_date_of_month_9',
       'date_of_last_rech_6', 'date_of_last_rech_7', 'date_of_last_rech_8',
       'date_of_last_rech_9']
no_missing_churn_df[date_attributes].head()
```

```python
no_missing_churn_df[date_attributes] = no_missing_churn_df[date_attributes].applymap(lambda x : pd.to_datetime(x, format = '%M/%d/%Y'))
```
